```python
import pandas as pd
from faker import Faker
import random
import string
from datetime import datetime
import string
from datetime import datetime, timedelta
```

```python
def generate_customer_data(num_records):
    customers = []
    for _ in range(num_records):
        customer_id = ''.join(random.choices(string.ascii_uppercase + string.digits, k=8))
        name = fake.name()
        date_of_birth = fake.date_of_birth(minimum_age=18, maximum_age=80)
        city = fake.city()
        state = fake.state()
        country = fake.country()
        postal_code = fake.postalcode()
        phone = fake.phone_number()
        email = fake.email()
        date_entered = fake.date_this_year()
        customers.append({
            'customer_id': customer_id,
            'name': name,
            'date_of_birth': date_of_birth,
            'city': city,
            'state': state,
            'country': country,
            'postal_code': postal_code,
            'phone': phone,
            'email': email,
            'date_entered': date_entered
        })
    return pd.DataFrame(customers)
```

```python
def generate_category_data():
    category = []
    pipe_categories = [
    "PVC Pipes", "CPVC Pipes", "UPVC Pipes", "PEX Pipes", "HDPE Pipes", 
    "LDPE Pipes", "PPR Pipes", "Galvanized Steel Pipes", "Stainless Steel Pipes", 
    "Copper Pipes", "Aluminum Pipes", "Concrete Pipes", "Cast Iron Pipes", 
    "Ductile Iron Pipes", "Polypropylene Pipes"]
    for i in (pipe_categories):
        category_id = fake.uuid4()
        CategoryName = i
        category.append({'category_id': category_id, 'CategoryName': CategoryName})
    return pd.DataFrame(category)
```

```python
# Generate product data
def generate_product_data(category_df):
    products = []
    pipe_names = [
    "FlowGuard Gold CPVC Pipe", "AquaPex PEX Pipe", "Duraflex PVC Pipe", "MegaFlow HDPE Pipe", 
    "ProGuard UPVC Pipe", "ThermoShield PPR Pipe", "TitanGalv Galvanized Steel Pipe", 
    "UltraPure Stainless Steel Pipe", "CopperFlex Pipe", "AlumaStrong Aluminum Pipe", 
    "ConcreteMax Pipe", "CastIron Classic Pipe", "DuctileForce Iron Pipe", 
    "PolyFlow Polypropylene Pipe", "FlexiDrain LDPE Pipe"
]

    for i in(pipe_names):
        product_id =''.join(random.choices(string.ascii_uppercase + string.digits, k=9))
        product_name = i
        category_id = random.choice(category_df['category_id'])
        price = random.randint(1000, 10000)
        products.append({'product_id': product_id, 'product_name': product_name, 'category_id': category_id,'price': price})
    return pd.DataFrame(products)
```

```python
def random_date(start_date, end_date):
    delta = end_date - start_date
    random_days = random.randint(0, delta.days)
    return start_date + timedelta(days=random_days)
```

```python
def random_delivery_date(order_date, min_days=1, max_days=10):
    delta_days = random.randint(min_days, max_days)
    return order_date + timedelta(days=delta_days)
```

```python
# Generate orders data
def generate_orders_data(num_records,customers_df, payments_df, shippers_df):
    Orders = []
    start_date = datetime(2021, 1, 1)
    end_date = datetime(2023, 12, 31)
    for _ in range(num_records):
        order_id = ''.join(random.choices(string.digits, k=5))
        customer_id = random.choice(customers_df['customer_id'])
        payment_id = random.choice(payments_df['payment_id'])
        order_date = random_date(start_date, end_date)
        shipper_id = random.choice(shippers_df['shipper_id'])
        delivery_date = random_delivery_date(order_date)
        total_order_amount = random.randint(5000, 20000)
        Orders.append({'order_id': order_id, 'customer_id': customer_id, 'payment_id': payment_id, 'order_date': order_date, 
                       'shipper_id':shipper_id, 'delivery_date': delivery_date, 'total_order_amount':total_order_amount})
    return pd.DataFrame(Orders)
```

```python
orders_df = generate_orders_data(100,customers_df, payments_df, shippers_df)
orders_df
```

```python
orders_df_csv = orders_df.to_csv('orders_df.csv', index=False)
```

```python
def generate_orderdetails_data(num_records,orders_df, product_df):
    OrderDetails = []
    for _ in range(num_records):
        OrderDetails_id = fake.uuid4()
        orderid = random.choice(orders_df['order_id'])
        product_id = random.choice(product_df['product_id'])
        quantity = random.randint(1,10)
        OrderDetails.append({'OrderDetails_id': OrderDetails_id,'orderid':orderid, 'product_id': product_id, 'quantity':quantity})
    return pd.DataFrame(OrderDetails)
```

```python
def generate_review_data(num_records, product_df, customers_df):
    review = []
    for _ in range(num_records):
        reviews_id = fake.uuid4()
        product_id = random.choice(product_df['product_id'])
        customer_id = random.choice(customers_df['customer_id'])
        rating = random.randint(1,5)
        review.append({'reviews_id': reviews_id, 'product_id': product_id, 'customer_id':customer_id, 'rating':rating})
    return pd.DataFrame(review)
```
